```python
@jit
def non_zero_idx(arr):
  idx = -1
  def _tmp(i, val):
    v = arr[i]
    idx = val
    return jax.lax.cond(v != 0, lambda: i, lambda: idx)
  return jax.lax.fori_loop(0, 10, _tmp, idx)

@jit
def set_action_cand(obs):
  cand = jnp.zeros(6, dtype=jnp.int8)
  def _set_idx(i, cand):
    arr = obs[i]
    idx = non_zero_idx(arr)
    cand = cand.at[i].set(idx)
    return cand
  cand = jax.lax.fori_loop(0, 5, _set_idx, cand)
  return cand


@jit
def _select_action(model, state, curr_player, obs):
  cand = set_action_cand(obs)
  return cand[0] # 候補手の中で常に一番indexが若いものを選択.



key = jax.random.PRNGKey(20)
curr_player, state = init(key)
while not state.terminated:
  print( jnp.where(state.legal_action_mask!=0)[0])
  print(state.playable_dice)
  print(state.turn)
  vis = Visualizer(state, scale=0.2)
  vis.show_svg(state)
  obs = observe(state, curr_player)
  action = jnp.where(state.legal_action_mask!=0)[0][0]
  curr_player, state, reward = step(state, action)
```
